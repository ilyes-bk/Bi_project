# Transform

```python
# Categorical Features
for column in car.select_dtypes(include='O').columns:
    print(f"Column: {column}")
    print(car[column].value_counts())
```

```python
# Nan Values
nan_count_per_column = df.isna().sum()

print("Number of NaN values in each column:")
print(nan_count_per_column)
```

```python
# Delete the Unnecessary Features
columns_to_delete = ['saler_type', 'owner_country','owner_street','owner_loaction','car_gearbox','owner_phone_number','car_displacement','car_weight','car_power','car_power','owner_location','ad_url','ad_title','ad_source','ad_creation_date','currency','data_type','ad_inactive']

# Drop the specified columns
df = df.drop(columns=columns_to_delete)
```

```python
nan_count_per_column = df.isna().sum()

print("Number of NaN values in each column:")
print(nan_count_per_column)
```

```python
# Specify the column with NaN values
column_with_nan = ['car_version','car_transmission','car_nb_doors','car_nb_seats','car_body','car_real_power']

# Drop rows with NaN values in the specified column
df = df.dropna(subset= column_with_nan)
```

```python
# Specify columns where you want to fill NaN values with the most frequent value
columns_to_fill = ['car_color', 'car_options']

# Fill NaN values with the most frequent value for the specified columns
df[columns_to_fill] = df[columns_to_fill].apply(lambda x: x.fillna(x.mode().iloc[0]))
```

```python
# Specify columns where you want to fill NaN values with the mean value
columns_to_fill = ['car_consumption', 'car_emission']

# Fill NaN values with the mean value for the specified columns
df[columns_to_fill] = df[columns_to_fill].apply(lambda x: x.fillna(x.mean()))
```

# PLOT

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Plot the distribution of car colors
plt.figure(figsize=(10, 6))
df['car_color'].value_counts().plot(kind='bar', color='skyblue')
plt.title('Distribution of Car Colors')
plt.xlabel('Car Color')
plt.ylabel('Count')
plt.show()
```

```python
# Plot a histogram of car prices
plt.figure(figsize=(10, 6))
sns.histplot(df['ad_price'], bins=100, color='skyblue', kde=True)
plt.title('Distribution of Car Prices')
plt.xlabel('Car Price')
plt.ylabel('Count')
plt.show()
```

```python
# Plot a bar plot of car brands
plt.figure(figsize=(14, 6))
sns.countplot(x='car_brand', data=df, palette='viridis')
plt.title('Distribution of Car Brands')
plt.xlabel('Car Brand')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.show()
```
